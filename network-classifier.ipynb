# 🧠 Neural Network Classifier

Demonstrates **Single Layer Neural Networks** and **MLP with Backpropagation** across four real-world classification scenarios:
- Student Performance Prediction
- Loan Approval System
- Weather Prediction
- Employee Performance Evaluation

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report

print('Libraries loaded!')
```

## Single Layer Neural Network (from scratch)

```python
class SingleLayerNN:
    def __init__(self, n_inputs, n_outputs, lr=0.1, epochs=1000):
        self.lr = lr
        self.epochs = epochs
        np.random.seed(42)
        self.W = np.random.randn(n_inputs, n_outputs) * 0.01
        self.b = np.zeros((1, n_outputs))

    def softmax(self, z):
        e = np.exp(z - np.max(z, axis=1, keepdims=True))
        return e / e.sum(axis=1, keepdims=True)

    def one_hot(self, y, n_classes):
        oh = np.zeros((len(y), n_classes))
        oh[np.arange(len(y)), y] = 1
        return oh

    def fit(self, X, y):
        n_classes = len(np.unique(y))
        Y = self.one_hot(y, n_classes)
        for epoch in range(self.epochs):
            A = self.softmax(X @ self.W + self.b)
            loss = -np.mean(np.sum(Y * np.log(A + 1e-9), axis=1))
            dZ = A - Y
            self.W -= self.lr * X.T @ dZ / len(X)
            self.b -= self.lr * dZ.mean(axis=0, keepdims=True)
            if (epoch + 1) % 200 == 0:
                print(f'  Epoch {epoch+1:>4d} | Loss: {loss:.4f}')

    def predict(self, X):
        return np.argmax(self.softmax(X @ self.W + self.b), axis=1)
```

---
## Scenario 1 – Student Performance

```python
X_student = np.array([[95,88],[70,60],[55,45],[92,85],[68,58]], dtype=float)
y_student_labels = ['Excellent','Average','Poor','Excellent','Average']

le_s = LabelEncoder(); y_s = le_s.fit_transform(y_student_labels)
sc_s = StandardScaler(); X_s = sc_s.fit_transform(X_student)

print('=== Single Layer NN ===')
slnn_s = SingleLayerNN(2, 3); slnn_s.fit(X_s, y_s)
preds = slnn_s.predict(X_s)
print('Predictions:', le_s.inverse_transform(preds))
print(f'Accuracy: {accuracy_score(y_s, preds)*100:.1f}%')

print('\n=== MLP (Backpropagation) ===')
mlp_s = MLPClassifier(hidden_layer_sizes=(8,4), max_iter=2000, random_state=42)
mlp_s.fit(X_s, y_s)
preds_m = mlp_s.predict(X_s)
print('Predictions:', le_s.inverse_transform(preds_m))
print(f'Accuracy: {accuracy_score(y_s, preds_m)*100:.1f}%')
print(classification_report(y_s, preds_m, target_names=le_s.classes_, zero_division=0))
```
